### Sentiment Analysis

```python
def detect_sentiment(text):
    return TextBlob(text).sentiment.polarity
```

```python
cdf=df[["Title"]]
```

```python
cdf.head()
```

```python
cdf['sentiment']=df["Title"].apply(detect_sentiment)
```

```python
cdf['Title']=cdf['Title'].str.lower()
cdf['Title']=cdf['Title'].str.replace('[^\w\s]','')
cdf['Title']=cdf['Title'].str.replace('\d+','')
cdf['Title']=cdf['Title'].str.replace('\n',' ').replace('\r',' ')
```

```python
cdf.sentiment.value_counts()
```

```python
def sentiment(sent):
    if sent<-0.02:
        return 2
    elif sent>0.02:
        return 1
    else:
        return 0
```

```python
cdf['sent']=cdf['sentiment'].apply(sentiment)
```

```python
cdf.head()
```

```python
sns.countplot(x=cdf.sent)
```

### LSTM Model

Bu bölümdeki kodlar "https://www.kaggle.com/code/johnybhiduri/next-word-prediction-using-lstm-on-news-headlines" adresinden alınmıştır.
LSTM konusu ilgi çekmiş olup öğrenmek amaçlı kullanılmıştır.

##### Tokenizing

```python
tokenizer = tf.keras.preprocessing.text.Tokenizer()
tokenizer.fit_on_texts(cdf['Title']) # giriş verisine göre tokinizer sınıfının ayarlanması
seq = tokenizer.texts_to_sequences(cdf['Title']) #veriler artık cümleler yerine sayısal değerler ifade edilmiş oldu.
```

```python
seq[:10]
```

```python
x = []
y = []
total_words_drop = 0
for i in seq:
    if len(i) > 1:
        for j in range(1, len(i)):
            x.append(i[:j])
            y.append(i[j])
            
    else : 
        total_words_drop +=1
print('Total Words Dropped : {}'.format(total_words_drop))
#Bazı cümlelerin [] şeklinde boş dizilere dönüşenler varsa çıkarıyoruz.
```

```python
x = tf.keras.preprocessing.sequence.pad_sequences(x)
```

```python
x.shape
```

```python
y = tf.keras.utils.to_categorical(y)
```

```python
y.shape
```

```python
vocab_size = len(tokenizer.word_index) + 1
vocab_size
```

```python
model = tf.keras.Sequential([tf.keras.layers.Embedding(vocab_size,49 ),
                            tf.keras.layers.LSTM(100, return_sequences = True),
                            tf.keras.layers.LSTM(100),
                            tf.keras.layers.Dense(100, activation = 'relu'),
                            tf.keras.layers.Dense(vocab_size, activation = 'softmax')])
```

```python
model.summary()
```

```python
model.compile(loss  = 'categorical_crossentropy',optimizer = 'adam',metrics = ['accuracy'])
```

```python
history = model.fit(x,y,epochs = 100,batch_size = 256,callbacks = tf.keras.callbacks.EarlyStopping(monitor = 'val_loss',
                                                               patience = 5,
                                                               restore_best_weights = True))
```

```python
vocab_array = np.array(list(tokenizer.word_index.keys()))
vocab_array
```

```python
def make_predictions(text, n_words):
    for i in range(n_words):
        text_tokenize = tokenizer.texts_to_sequences([text])
        text_padding  = tf.keras.preprocessing.sequence.pad_sequences(text_tokenize, maxlen = 49)
        prediction = np.squeeze(np.argmax(model.predict(text_padding), axis = -1))
        prediction = str(vocab_array[prediction - 1])
        text += " " + prediction
    return text
```

```python
#Prediction
make_predictions('san francisco',5)
```
